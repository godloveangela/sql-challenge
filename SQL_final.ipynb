```python
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg2://postgres:postgres@150.158.149.214:54321/postgres')
connection = engine.connect()
```

```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
# List the following details of each employee: employee number, last name, first name, sex, and salary.
pd.read_sql_query('SELECT employees.emp_no, employees.last_name, employees.first_name, employees.sex, salaries.salary FROM employees, salaries WHERE employees.emp_no = salaries.emp_no LIMIT 50',connection)
```

```python
# List first name, last name, and hire date for employees who were hired in 1986.
sql_query = "SELECT employees.first_name, employees.last_name, employees.hire_date FROM employees WHERE employees.hire_date BETWEEN '1986-01-01' AND '1986-12-31'"
import sqlalchemy
sql_query = sqlalchemy.text(sql_query)
engine.execute(sql_query).fetchmany(50)
```

```python
# List the manager of each department with the following information: department number, department name, the manager's employee number, last name, first name.
pd.read_sql_query("select departments.dept_no, departments.dept_name, dept_manager.emp_no, employees.last_name, employees.first_name from departments, dept_manager, employees where departments.dept_no = dept_manager.dept_no and dept_manager.emp_no = employees.emp_no LIMIT 50",connection)
```

```python
# List the department of each employee with the following information: employee number, last name, first name, and department name.
sql_query = "SELECT employees.emp_no, employees.last_name, employees.first_name, departments.dept_name from employees, dept_emp, departments WHERE (employees.emp_no = dept_emp.emp_no) AND (dept_emp.dept_no = departments.dept_no) AND (departments.dept_name = 'Sales') LIMIT 50"

pd.read_sql_query(sql_query,connection)
```

```python
# List first name, last name, and sex for employees whose first name is "Hercules" and last names begin with "B."
import sqlalchemy as sa
sql_query = ("SELECT first_name, last_name, sex from employees WHERE (first_name = 'Hercules') AND (last_name LIKE 'B%')")
sql_query = sa.text(sql_query)
pd.read_sql_query(sql_query, connection)
```

```python
# List all employees in the Sales department, including their employee number, last name, first name, and department name.
sql_query = "SELECT employees.emp_no, employees.last_name, employees.first_name, departments.dept_name from employees, dept_emp, departments WHERE (employees.emp_no = dept_emp.emp_no) AND (dept_emp.dept_no = departments.dept_no) AND (departments.dept_name = 'Sales') LIMIT 50"
sql_query = sa.text(sql_query)
engine.execute(sql_query).fetchmany(50)
```

```python
# List all employees in the Sales and Development departments, including their employee number, last name, first name, and department name.
sql_query = "SELECT employees.emp_no, employees.last_name, employees.first_name, departments.dept_name from employees, dept_emp, departments WHERE (employees.emp_no = dept_emp.emp_no) AND (dept_emp.dept_no = departments.dept_no) AND (departments.dept_name = 'Sales' OR departments.dept_name = 'Development') LIMIT 50"
sql_query = sa.text(sql_query)
engine.execute(sql_query).fetchmany(50)
```

```python
# In descending order, list the frequency count of employee last names, i.e., how many employees share each last name.
sql_query = "SELECT last_name, COUNT(last_name) FROM employees GROUP BY last_name ORDER BY COUNT(last_name) DESC LIMIT 50"
sql_query = sa.text(sql_query)
engine.execute(sql_query).fetchmany(50)
```

```python
# Create a histogram to visualise the most common salary ranges for employees.
hist = pd.read_sql_query('SELECT * FROM salaries', connection)
hist.plot.hist(bins=10)
```

```python
# Create a bar chart of average salary by title.
sql_query = 'SELECT title, "count"(title), "avg"("salary") FROM employees, title, salaries WHERE (employees.emp_title_id = title.title_id) and (employees.emp_no = salaries.emp_no) GROUP BY title.title LIMIT 1000'
sql_query = sa.text(sql_query)


title = []
avg_salary = []
for i in engine.execute(sql_query).fetchall():
    title.append(i[0])
    avg_salary.append(i[2])

import matplotlib.pyplot as plt
plt.bar(title, avg_salary)
```
